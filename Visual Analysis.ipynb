```python
# Part 1.1: Click Rate Data Visualization

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# Set seaborn theme for professional aesthetics
sns.set_theme(style="whitegrid", context="notebook", palette="muted")

# Load the CSV file (update the file path as needed)
df = pd.read_csv("/Users/parker.pape/Downloads/Thumbnail_Colorfulness_Analysis_Final.csv")

# Initialize the plot
plt.figure(figsize=(10, 6))

# Create a scatterplot with a regression line
scatter_plot = sns.regplot(
    data=df,
    x="click_rate",
    y="Colorfulness Score",
    scatter_kws={"s": 60, "alpha": 0.7},
    line_kws={"color": "red", "lw": 2},
    ci=None
)

# Set title and axis labels
plt.title("Correlation between Click Rate Percentage and Colorfulness Score", fontsize=14, weight='bold')
plt.xlabel("Click Rate Percentage", fontsize=12)
plt.ylabel("Colorfulness Score", fontsize=12)

# Format x-axis as percentages if needed
scatter_plot.xaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0))  # remove if values are already in percent form

# Optional: Add a grid and tight layout for cleanliness
plt.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Part 1.2: Completion Rate Data Visualization

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# Set seaborn theme for professional aesthetics
sns.set_theme(style="whitegrid", context="notebook", palette="muted")

# Load the CSV file (update the file path as needed)
df = pd.read_csv("/Users/parker.pape/Downloads/Thumbnail_Colorfulness_Analysis_Final.csv")

# Initialize the plot
plt.figure(figsize=(10, 6))

# Create a scatterplot with a regression line
scatter_plot = sns.regplot(
    data=df,
    x="completion_rate",
    y="Colorfulness Score",
    scatter_kws={"s": 60, "alpha": 0.7},
    line_kws={"color": "red", "lw": 2},
    ci=None
)

# Set title and axis labels
plt.title("Correlation between Completion Percentage and Colorfulness Score", fontsize=14, weight='bold')
plt.xlabel("Completion Percentage", fontsize=12)
plt.ylabel("Colorfulness Score", fontsize=12)

# Format x-axis as percentages if needed
scatter_plot.xaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0))  # remove if values are already in percent form

# Optional: Add a grid and tight layout for cleanliness
plt.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Part 2: Statistical Calculation

from scipy.stats import pearsonr

# Click Rate Stats
# Remove missing data
df = df.dropna(subset=['Colorfulness Score'])

# Extract variables
x = df["Colorfulness Score"]
y = df["click_rate"]

# Calculate Pearson correlation
r_value, p_value = pearsonr(x, y)
r_squared = r_value ** 2

# Display the results
print("Click Rate Stats:")
print(f"Pearson correlation coefficient (r): {r_value:.4f}")
print(f"Coefficient of determination (r): {r_squared:.4f}")
print(f"P-value: {p_value:.4e}")
print("Entries:",len(x), "\n")



# Completion Rate Stats
# Extract variables
x = df["Colorfulness Score"]
y = df["completion_rate"]

# Calculate Pearson correlation
r_value, p_value = pearsonr(x, y)
r_squared = r_value ** 2

# Display the results
print("Completion Rate Stats:")
print(f"Pearson correlation coefficient (r): {r_value:.4f}")
print(f"Coefficient of determination (r): {r_squared:.4f}")
print(f"P-value: {p_value:.4e}")
print(f"Entries: {len(x)}")

# Completion Rate Stats
```
